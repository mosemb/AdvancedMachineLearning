# Loading dataset and Applying Transformations to use.

## Trial 1: Taking 3 middle slices from each stack.

```python
base_path = '/content/gdrive/My Drive/Dataset/MRNet-v1.0'
trans_path = '/content/gdrive/My Drive/Dataset/Transformations'

for data_type in ('train','valid'):
  for plane in ('axial','coronal','sagittal'):
    orig_path = base_path+'/'+data_type+'/'+plane+'/'
    new_path = trans_path+'/mid3_'+data_type+'_'+plane+'.npy'
    transform_data.apply_three_slice_trans(orig_path, new_path)
```

## Interpolating to 24 slices per exam

```python
!mkdir /content/gdrive/My\ Drive/Transformations
```

```python
base_path = '/content/gdrive/My Drive/Dataset/MRNet-v1.0'
trans_path = '/content/gdrive/My Drive/Transformations'
slices=24
for data_type in ('train','valid'):
  for plane in ('axial','coronal','sagittal'):
    orig_path = base_path+'/'+data_type+'/'+plane+'/'
    new_path = trans_path+'/interpolated24'+'_'+data_type+'_'+plane+'.npy'
    transform_data.apply_interpolation_transformation(orig_path, new_path,slices)
```

```python
base_path = '/content/gdrive/My Drive/Dataset/MRNet-v1.0'
trans_path = '/content/gdrive/My Drive/Dataset/Transformations/Five_Class_labels'

for data_type in ('train','valid'):
  transform_data.apply_five_class_labels(base_path+'/'+data_type+'-abnormal.csv',
                                        base_path+'/'+data_type+'-acl.csv',
                                        base_path+'/'+data_type+'-meniscus.csv',
                                        trans_path+'/'+data_type+'-labels.npy')
```
